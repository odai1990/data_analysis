```python
import pandas as pd
import numpy as np
df = pd.read_csv('./vgsales.csv')
```

# 1.Which company is the most common video game publisher?

```python
df['Publisher'].describe().top
```

# 2- What's the most common platform?

```python
df['Platform'].describe().top
```

# 3- What about the most common genre?

```python
df['Genre'].describe().top
```

# 4- What are the top 20 highest grossing games?

```python
df.loc[0:19][['Name','Global_Sales']]
```

# 5- For North American video game sales, what's the median?

```python
df['NA_Sales'].median()
```

# 6- For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python
(df['NA_Sales'].iloc[0]-df['NA_Sales'].mean())/df['NA_Sales'].std()
```

# 7- The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python
data=df.groupby(['Platform','Global_Sales'])
```

```python
# data.agg({'Global_Sales':'mean'})
all=df[df['Platform']=='Wii']['Global_Sales'].mean()
wii=df['Global_Sales'].mean()
```

```python
(wii,all)
```

# 8-Top 10 Nintendo Games?

```python
data_n = df[df['Publisher'] == 'Nintendo']
data_n.sort_values(by='Global_Sales', ascending=False)
data_n[:10]
```

# 9- Most Popular Shooters?

```python
data_s = df[df['Genre'] == 'Shooter']
data_s.sort_values(by='Global_Sales', ascending=False)
data_s[:20]
```

# 10- Top games in 2010?

```python
data_t = df[df['Year'] == 2010]
data_t.sort_values(by='Global_Sales', ascending=False)
data_t[:20]
```

```python

def test():

    def assert_equal(actual,expected):
        assert actual == expected, f"Expected {expected} but got {actual}"

    assert_equal(df['Publisher'].describe().top, 'Electronic Arts')
    assert_equal(df['Platform'].describe().top, 'DS')
    assert_equal(df['Genre'].describe().top,'Action')
    assert_equal(df.loc[0:19][['Name','Global_Sales']].iloc[7]['Name'], 'Wii Play')
    assert_equal(df['NA_Sales'].median(), 0.08)
    assert_equal((df['NA_Sales'].iloc[0]-df['NA_Sales'].mean())/df['NA_Sales'].std(), 50.47898767479108)
    assert_equal((wii,all), (0.5374406555006628, 0.6994037735849057))
    assert_equal(data_n.iloc[9]['Platform'], 'NES')
    assert_equal(data_s.iloc[10]['Publisher'], 'Activision')
    assert_equal(data_t.iloc[19]['Genre'], 'Misc')
    print("Success!!!")

test()
```
